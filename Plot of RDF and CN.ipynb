```python
import numpy as np
import matplotlib.pyplot as plt
def load_xvg(filename):
    data = []
    with open(filename, 'r') as f:
        for line in f:
            if line.startswith(('#', '@')):
                continue
            parts = line.split()
            if not parts:
                continue
            data.append([float(x) for x in parts])
    return np.array(data)


rdf = load_xvg('rdf_NaCl_shell.xvg')
cn  = load_xvg('cn_NaCl_cmumd.xvg')

r_rdf = rdf[:, 0]
g_rdf = rdf[:, 1]

r_cn  = cn[:, 0]
cn_r  = cn[:, 1]

# RDF
plt.figure(figsize=(6,4))
plt.plot(r_rdf, g_rdf)
plt.xlabel('r (nm)')
plt.ylabel('g_NaCl(r)')
plt.title('RDF: NaCl (CmuMD)')
plt.tight_layout()
plt.savefig('rdf_NaCl_cmumd.png', dpi=600)

# cumulative coordination number
plt.figure(figsize=(6,4))
plt.plot(r_cn, cn_r)
plt.xlabel('r (nm)')
plt.ylabel('CN(r)')
plt.title('Cumulative coordination number: NaCl (CmuMD)')
plt.tight_layout()
plt.savefig('cn_NaCl_cmumd.png', dpi=600)

plt.show()
```
