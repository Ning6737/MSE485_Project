```python
import numpy as np

data = []

with open("cmumd_shell_pre.out") as f:
    for line in f:
        if line.startswith(('#', '@')):
            continue
        parts = line.split()
        if len(parts) >= 7:
            vals = [float(x) for x in parts[:7]]
            data.append(vals)

data = np.array(data)
time      = data[:,0]
cn_na_Rgr = data[:,1]
cn_cl_Rgr = data[:,2]
cn_na_Rcr = data[:,3]
cn_cl_Rcr = data[:,4]
na_CR     = data[:,5]
cl_CR     = data[:,6]

print("na_CR mean =", na_CR.mean())
print("cl_CR mean =", cl_CR.mean())
print("na_CR std  =", na_CR.std())
print("cl_CR std  =", cl_CR.std())
```
